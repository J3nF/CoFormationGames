```julia
using GraphRecipes, Plots
using Graphs
```

```julia
g = wheel_graph(3)
edgelabel_dict = Dict((1,2) => "0", (2,3) => "1", (1,3) => "1")
graphplot(g,
    x = [0, 1, 0.5],
    y = [0, 0, 1],
    names=[0,0,1], # node labels are opinions the nodes hold
    edgelabel=edgelabel_dict, # edge labels are opinion differences
    curves=false,
    nodeshape=:circle,
    nodecolor=:lightblue,
    markersize=0.15,
    fontsize=16,
)

savefig("0_triangle.png")
```

```julia
g = path_graph(3)
edgelabel_dict = Dict((1,2) => "0", (2,3) => "1")
graphplot(g,
    x = [0, 1, 2], # position nodes in a line
    y = [0, 0, 0],
    names=[0,0,1], # node labels are opinions the nodes hold
    edgelabel=edgelabel_dict, # edge labels are opinion differences
    curves=false,
    nodeshape=:circle,
    nodecolor=:lightblue,
    markersize=0.2,
    fontsize=16,
)
savefig("0_outchain.png")
```
